```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(0, 2, 10)
y  = np.linspace(0,3,10)
plt.plot(x, y)
x = np.linspace(0, 2, 10)
y  = np.linspace(0,0,10)
plt.plot(x, y)
x = np.linspace(2, 2, 10)
y  = np.linspace(0,3,10)
plt.plot(x, y)
```
